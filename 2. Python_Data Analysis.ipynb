# `Payment Success/Failure Rates by Gender`

```python
# Merging gender data from 'customer' table with 'transactions' table
payment_status_merged = pd.merge(transactions, customer[['customer_id', 'gender']], on='customer_id', how='left')

# Counting the number of success and failure statuses by gender
payment_status_count = payment_status_merged.groupby(['gender', 'payment_status'])['payment_status'].count().unstack(fill_value=0)

# Calculating the total counts and the percentage
payment_status_percentage = round(payment_status_count.div(payment_status_count.sum(axis=1), axis=0) * 100,2)

# Combining counts and percentages into a single DataFrame
result = pd.DataFrame({
    'Count': payment_status_count.stack(),
    'Percentage': payment_status_percentage.stack()
}).unstack()

# Displaying the results
print("Payment Status Counts and Percentages by Gender:")
result
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
payment_status_count.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], ax=ax)

# Loop through each segment and add percentage labels
for i, (gender, counts) in enumerate(payment_status_count.iterrows()):
    total = counts.sum()
    for j, (status, count) in enumerate(counts.items()):
        if count > 0:
            percentage = payment_status_percentage.loc[gender, status]
            ax.text(i, counts.cumsum()[status] - count / 2, f"{percentage:.1f}%", ha='center', va='center', color="black")

plt.title('Payment Success and Failure Rates by Gender with Percentages')
plt.xlabel('Gender')
plt.ylabel('Number of Transactions')
plt.legend(title='Payment Status', loc='upper left')
plt.show()
```

### <center>COMMENT
<center>  We Can see that failure percentage for both female and male are almost same

# `Seasonal Trends by SALE`

```python
# Extract month from transaction date
transactions['month'] = transactions['created_at'].dt.month_name()

# Define a function to map each month to a season
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'April']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Fall'

# Apply the function to create a 'season' column
transactions['season'] = transactions['month'].apply(get_season)
```

```python
transactions.head()
```

```python
# Group by season and count the number of transactions
seasonal_trends = transactions.groupby('season')['total_amount_USD'].count().reset_index().sort_values('total_amount_USD',
                                                                                ascending=False).reset_index(drop=True)
seasonal_trends


# Plotting the seasonal trends
plt.figure(figsize=(10, 6))
plt.bar(seasonal_trends['season'], seasonal_trends['total_amount_USD'], color=['skyblue', 'salmon', 'lightgreen', 'orange'])
plt.xlabel('season')
plt.ylabel('Total Transactions')
plt.title('Transactions by Season')
plt.show()
```

# `Finding Peak Shopping Times of the Day`

```python
# Extract hour from the transaction time
transactions['hour'] = transactions['created_at'].dt.hour
peak_shopping_times = transactions.groupby('hour')['created_at'].count().reset_index(name='transaction_count')

# categorize the hour
def categorize_hour(hour):
    if hour >= 5 and hour < 12 : return 'Morning'
    elif hour >= 12 and hour < 17 : return 'Afternoon'
    elif hour >= 17 and hour < 21 : return 'Evening'
    else:
        return 'Night'
    
    
peak_shopping_times['time_category'] = peak_shopping_times['hour'].apply(categorize_hour)
peak_shopping_times = peak_shopping_times.groupby('time_category')['transaction_count'].sum().reset_index()
peak_shopping_times['percentage'] = round((peak_shopping_times['transaction_count'] / peak_shopping_times['transaction_count'].sum()) * 100,2)   

peak_shopping_times.sort_values('percentage', ascending = False).reset_index(drop=True)
```

#### <center>COMMENT :
<center> Customers are tend to do their shopping more at Night time.
